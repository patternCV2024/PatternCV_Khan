<h2>Побудова дерева ухвалення рішень мовою python</h2>

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree

# Функція для створення сітки
def get_grid(data):
    x_min, x_max = data[:, 0].min() - 1, data[:, 0].max() + 1
    y_min, y_max = data[:, 1].min() - 1, data[:, 1].max() + 1
    return np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))

# Ваші дані
x_train_6 = np.array([[49, 21], [5, 5], [37, 32], [21, 25], [34, 28], [44, 35], [39, 41], [17, 45], [31, 24]])
y_train_6 = np.array([-1, 1, 1, -1, -1, -1, 1, -1, 1])

# Ініціалізація та навчання моделі дерева ухвалення рішень
clf_tree = DecisionTreeClassifier(criterion='entropy', max_depth=3, random_state=5)
clf_tree.fit(x_train_6, y_train_6)

# Отримання прогнозів та відображення сітки та точок даних
xx, yy = get_grid(x_train_6)
predicted = clf_tree.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
plt.pcolormesh(xx, yy, predicted, cmap='spring', shading='auto')
plt.scatter(x_train_6[:, 0], x_train_6[:, 1], c=y_train_6, s=50, cmap='spring', edgecolors='black', linewidth=1.5)
plt.show()

# Візуалізація дерева
plot_tree(clf_tree)
```

3. **Критерії якості для побудови дерев ухвалення рішень:**

   - Ентропія: міра невизначеності в системі, використовується для вимірювання того, наскільки добре вузли дерева розділяють дані за допомогою атрибутів.

   - Джині: міра чистоти чи гомогенності вузла дерева, використовується для визначення оптимального розділення даних.
   - Помилка класифікації: вимірюється точність класифікації моделі на основі випадкового вибору.

4. **Робота алгоритму ID3 у контексті побудови дерев класифікації:**

   - ID3 (Iterative Dichotomiser 3) є одним з перших алгоритмів для побудови дерев класифікації.
   - Він працює на принципі жадібного підходу, рекурсивно розділяючи набір даних на більші частини шляхом вибору найбільш інформативного атрибута на кожному кроці.
   - На кожному кроці алгоритм обирає атрибут, який найкраще розділяє набір даних, зазвичай за допомогою критерію інформаційного приросту або ентропії.
   - Процес триває до тих пір, поки всі дані не будуть розділені на чисті класи або досягнута максимальна глибина дерева.
   - Результатом є дерево рішень, яке може бути використане для класифікації нових даних за їхніми атрибутами.
